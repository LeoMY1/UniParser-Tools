```python
import base64
import json

from IPython.display import HTML, display

from uniparser_tools.utils.visualize_together import plotly_pdf_results
```

```python
json_path = "./results/aeb129da-4a36-4627-9cb2-c721def237f0.json"
pdf_path = "./tasks/aeb129da-4a36-4627-9cb2-c721def237f0.pdf"

with open(json_path, "r") as f:
    pages_dict = json.load(f)

# 选定页数，在 notebook 中运行,不支持点击页面交互，请自行 编写 server代码
html = plotly_pdf_results(pages_dict, pdf_path, pages=[0])

# 将 HTML 包装在固定尺寸的容器中
# 可以根据需要调整宽度和高度（单位：px 或 %）
width = "100%"  # 例如: "800px" 或 "100%"
height = "800px"  # 例如: "600px" 或 "800px"

# 使用 iframe（推荐，完全隔离，避免样式冲突）
html_bytes = html.encode('utf-8')
html_base64 = base64.b64encode(html_bytes).decode('utf-8')
wrapped_html = f"""
<div style="width: {width}; height: {height}; border: 1px solid #ddd; overflow: hidden; box-sizing: border-box;">
    <iframe src="data:text/html;base64,{html_base64}" 
            style="width: 100%; height: 100%; border: none; display: block;">
    </iframe>
</div>
"""

display(HTML(wrapped_html))
```
